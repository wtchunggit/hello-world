```python
from IPython.display import display
import datetime
import pandas as pd

def finish_time_to_timedelta(time):
    try:
        minutes, remainder = time.split(':')
        seconds, centiseconds = remainder.split('.')
        assert len(centiseconds) == 2
        return datetime.timedelta(seconds=float(seconds), milliseconds=float(centiseconds + '0'), minutes=float(minutes))
    except Exception as e:
        print(e, time)

def clean_race(df):
    # Returns a deep copy of races dataframe w/o rows containing missing data.
    # INPUTS:
    # `df` is created by reading *_races.csv
    
    #df = df[df['Pla'] != 'WV']
    df['Pla'] = pd.to_numeric(df['Pla'], errors='coerce')
    df = df[df['Pla'].notna()]
    df = df.copy()
    
    df['Win Odds'] = pd.to_numeric(df['Win Odds'], errors='coerce')
    df = df[df['Win Odds'].notna()]
    df = df.copy()
    
    df['Act. Wt.'] = pd.to_numeric(df['Act. Wt.'], errors='coerce')
    df['Declar. Horse Wt'] = pd.to_numeric(df['Declar. Horse Wt'], errors='coerce')
    df = df[df['Act. Wt.'].notna() & 
            df['Declar. Horse Wt'].notna() ]
    df = df[~df['Finish Time'].str.contains('-')]
    df = df.copy()
    
    return df


def add_columns_race(df):
    # Adds columns and formats columns.
    # INPUTS:
    # `df` is created by reading *_races.csv
    
    # Add columns  weight, carry weight (%), distance (m), average speed (m/s):
    # Add weight:
    df = df.astype(dtype = {'Declar. Horse Wt': 'float', 'Act. Wt.': 'float'}) #Typecast columns.
    df['wt'] = df['Act. Wt.'] + df['Declar. Horse Wt']  #make new column for total wt
    # Add carried weight (%):
    df['wt carry (%)'] = df['Act. Wt.'].divide(df['Declar. Horse Wt']) * 100.0
    # Add distance (m):
    df['distance (m)'] = df['Class & Distance'].str.split('-').apply(lambda L: L[1]).str.strip().apply(lambda s: s[:-1]) 
    df = df.astype(dtype = {'distance (m)': 'float'}) #Typecast column.
    # Add average speed (m/s):
    s = df['Finish Time'].apply(finish_time_to_timedelta) #Convert timing to timedelta object.
    df['Finish Time (s)'] = s.apply(lambda x: x.total_seconds())
    df['ave speed (m/s)'] = df['distance (m)'].divide(df['Finish Time (s)'])

    # Format some columns:
    df['Going'] = df['Going'].astype('category').cat.codes
    df = df.astype(dtype = {'Dr.': 'int'}) #Typecast column.
    df = df.astype(dtype = {'Pla': 'int'})
    df = df.astype(dtype = {'Win Odds': 'float'})
    
    return df

# Read race files
import glob
import pandas as pd
filepaths = [p for p in glob.glob('./data/race/*_HKJC_races.csv')]
dfs = (pd.read_csv(f) for f in filepaths)
df = pd.concat(dfs, ignore_index=True)
print(len(df))

# Remove rows with missing data
df = clean_race(df)
print(len(df))

# Add columns, format some columns.
df = add_columns_race(df)
pd.set_option('display.max_columns', None)  
df.head()

# Encode 'Course'
c = df['Course'].astype('category')
cat_course = dict(enumerate(c.cat.categories))
print(cat_course)
df['Course code'] = c.cat.codes
# Encode 'Class'
# TODO: What's the difference between these: '4 Year Olds ', 'Class 1 ', 'Class 2 ', 'Class 3 ', 'Class 4 ', 'Class 5 ', 'Griffin Race ', 'Group One ', 'Group Three ', 'Group Two ' ?
# df['Class'] = df['Class & Distance'].str.split('-').apply(lambda L: L[0]).str.split().apply(lambda L: L[1])
#df = df.astype(dtype = {'Class': 'int'})
df['Class'] = df['Class & Distance'].str.split('-').apply(lambda L: L[0])
c = df['Class'].astype('category')
cat_class = dict(enumerate(c.cat.categories))
print(cat_class)
df['Class code'] = c.cat.codes

df
```
